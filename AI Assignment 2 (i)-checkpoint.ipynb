```python
import matplotlib.pyplot as plt
import networkx as nx
import random


# Parameters
number_of_nodes = 100  # Number of nodes
probability = 0.1  # Probability of edge creation
colour_names = [
    "red", "green", "blue", "yellow", "orange", 
    "purple", "pink", "lime", "cyan", "magenta"
]

# Generate a random graph (Erdős–Rényi model)
graph = nx.erdos_renyi_graph(number_of_nodes, probability)
```

```python
# Function to ensure at least min_degree connections for each node
def ensure_minimum_degree(g, minimum_degree=2):
    for node in g.nodes():
        while g.degree(node) < minimum_degree:
            possible_nodes = set(g.nodes()) - set(nx.neighbors(g, node)) - {node}
            if not possible_nodes:
                break
            new_neighbour = random.choice(list(possible_nodes))
            g.add_edge(node, new_neighbour)

ensure_minimum_degree(graph)
    
# Randomly assign colours to each node
colour_map = {node: random.choice(colour_names) for node in graph.nodes}
```

```python
# Decentralised colouring algorithm
def decentralised_colouring(g, colour_map, max_iterations=100):
    conflict_history = []
    for _ in range(max_iterations):
        conflicts = 0
        for edge in g.edges:
            if colour_map[edge[0]] == colour_map[edge[1]]:
                conflicts += 1
        conflict_history.append(conflicts)
        if conflicts == 0:
            break
        for node in g.nodes:
            neighbour_colours = [colour_map[neighbour] for neighbour in g.neighbors(node)]
            available_colours = list(set(colour_names) - set(neighbour_colours))
            if colour_map[node] in neighbour_colours and available_colours:
                colour_map[node] = random.choice(available_colours)
    return conflict_history
```

```python
# Run the decentralised colouring experiment
conflict_history = decentralised_colouring(graph, colour_map)




# Plot the number of conflicts over time
plt.figure(figsize=(10, 6))
plt.plot(range(len(conflict_history)), conflict_history, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Number of Conflicts')
plt.title('Graph Colouring Conflicts Over Time with Minimum 2 Connections per Node')
plt.grid(True)
plt.show()

# Display the graph with coloured nodes
plt.figure(figsize=(12, 8))
colours = [colour_map[node] for node in graph.nodes]
nx.draw(graph, node_color=colours, with_labels=True, node_size=500)
plt.title('Graph Colouring Visualisation with Minimum 2 Connections per Node')
plt.show()

# Print out all the node connections(edges)
connections = [f"Node {edge[0]} -> Node {edge[1]}" for edge in graph.edges]
for connection in connections:
    print(connection)

# Output the colour of each node as a dictionary
colour_map
```
