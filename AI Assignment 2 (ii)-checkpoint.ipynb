```python
import matplotlib.pyplot as plt
import networkx as nx
import random
import numpy as np
import time

# Experiment parameters
probability = 0.1  # Fixed probability of edge creation
colour_names = ["red", "green", "blue", "yellow", "orange", 
                "purple", "pink", "lime", "cyan", "magenta"]  # Colours available
node_sizes = np.arange(1, 1001, 50)  # Node sizes to test
probability = 0.1  # Probability for edge creation
iterations_results = {}  # To store the iteration results
time_results = {}  # To store the time results 
```

```python
# Function to run the decentralized graph coloring algorithm
def decentralised_colouring(graph, colours):
    iterations = 0
    while True:
        iterations += 1
        conflict = False
        for node in graph.nodes():
            neighbour_colours = {colours[neighbour] for neighbour in graph.neighbors(node)}
            if colours[node] in neighbour_colours:
                available_colours = list(set(colour_names) - neighbour_colours)
                if available_colours:  # Only if there are available colours
                    colours[node] = random.choice(available_colours)
                    conflict = True
        if not conflict:  # No conflicts, break out of the loop
            break
    return iterations
```

```python
# Main experiment loop
for n in node_sizes:
    num_colours = int(np.ceil(np.sqrt(n)))  # Number of colours based on the graph size
    colour_palette = [f"Colour {i}" for i in range(1, num_colours + 1)]

    iterations_list = []  # Store the iterations for each run to take the average
    times = []  # Store the times for each run to take the average
    for i in range(10):  # Repeat the experiment 10 times for averaging
        G = nx.erdos_renyi_graph(n, probability)
        initial_colours = {node: random.choice(colour_palette) for node in G.nodes()}

        start_time = time.time()  # Time measurement start
        iterations = decentralised_colouring(G, initial_colours)
        end_time = time.time()  # Time measurement end

        iterations_list.append(iterations)  # Add the number of iterations for this run
        times.append(end_time - start_time)  # Calculate the time taken for this run

    iterations_results[n] = np.mean(iterations_list)  # Store the average iterations for this node size
    time_results[n] = np.mean(times)  # Store the average time for this node size

# Plotting the results 
plt.figure(figsize=(12, 7))
plt.plot(list(iterations_results.keys()), list(iterations_results.values()), marker='o')
plt.xlabel('Number of Nodes')
plt.ylabel('Iterations ')
plt.title('Impact of Number of Nodes on Number of Iterations Needed')
plt.grid(True)
plt.show()

# Plotting the time results
plt.figure(figsize=(12, 6))
plt.plot(list(time_results.keys()), list(time_results.values()), marker='o', color='red')
plt.xlabel('Number of Nodes')
plt.ylabel('Average Time Taken (seconds)')
plt.title('Impact of Number of Nodes on Time Taken')
plt.grid(True)
plt.show()
```
